```python
###################
## Run this cell ##
###################
import pandas as pd
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()

df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['Y'] = diabetes.target

print("Classification or Regression?")
print("-----------------------------")
print(diabetes.DESCR)
```

```python
"""
Training set to train model.
Test set for final evaluation ( to estimate generalization error )
"""
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop(['Y'], axis=1), df['Y'],
                                                    test_size=0.2, random_state=2021)

print(y_train.head())
x_train.head()
```

# Example exercise

```python
# 1. Import the necessary model.
from sklearn.linear_model import LinearRegression # after write this code, do ctrl+enter

# 2. Define your model.
lr = LinearRegression()

# 3. Fit the model to your data.
lr.fit(x_train, y_train)

# 4. Make predictions using the trained model.
y_pred = lr.predict(x_test)
```
