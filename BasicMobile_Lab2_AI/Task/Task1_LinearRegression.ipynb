# Q1. Split the df into training set & test set

1. x : all columns in df except 'Y'
2. y : the column 'Y' in df (df['Y'])
3. variable names :
    * x_train, y_train
    * x_test, y_test
4. train : test = 8 : 2
5. randomstate : 2024

Question : Why we need to prepare test set?

````python
####################
## Your code here ##
####################

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes

# 데이터셋 불러오기
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['Y'] = diabetes.target

# 데이터 분할
X = df.drop(columns='Y')
y = df['Y']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2024)

# 테스트 셋이 필요한 이유: 모델이 새로운 데이터를 얼마나 잘 예측할 수 있는지 평가하기 위함



````

# Q2. Train linear regression model

1. declare your model as lr

````python
####################
## Your code here ##
####################

from sklearn.linear_model import LinearRegression

# 선형 회귀 모델 선언 및 학습
lr = LinearRegression()
lr.fit(x_train, y_train)


# 실제 의료 데이터에서는 회귀와 분류 문제를 혼합하여 사용할 때도 있음
````

# Q3. Make a prediction

````python
####################
## Your code here ##
####################

# 훈련 세트와 테스트 세트에 대한 예측 수행
y_pred_train = lr.predict(x_train)
y_pred_test = lr.predict(x_test)


# 데이터셋의 출처와 사용 목적을 확인하는 것이 전제조
````

# Q5. Complete the equation of the linear regression model

\begin{align}
Y = \beta_{10} &\ + \beta_0*AGE + \beta_1*SEX + \beta_2*BMI + \beta_3*BP \\
&+ \beta_4*S1 + \beta_5*S2 + \beta_6*S3 + \beta_7*S4 + \beta_8*S5 + \beta_9*S6
\end{align}

* print $ \beta_0 $ ~ $ \beta_{10} $ with feature(column) name
* example
```
beta_0 for AGE : 21
beta_1 for SEX : - 12
~~~
beta_10 for intercept : -5
```



````python
####################
## Your code here ##
####################

# 회귀 계수와 절편 출력
coefficients = pd.Series(lr.coef_, index=X.columns)
intercept = lr.intercept_

for i, feature in enumerate(coefficients.index):
    print(f"beta_{i} for {feature}: {coefficients[feature]}")
print(f"beta_intercept: {intercept}")


# 각 피처의 회귀 계수를 출력하고, 피처별로 타겟 변수(Y)에 미치는 영향을 확인
````

# Q6. Analyze the effect of 'AGE' (average average number of rooms per dwelling)

**assumption1 : every other features are fixed.**<br>
**assumption2 : use training set to anlayze.**
1. How does the 'Y' change when 'AGE' increases by 1 ?
2. What is the expected change in the 'Y' due to one standard deviation increase in 'AGE'?
    * hint : beta_0 * std('AGE')
3. What is the expected change in the 'Y' due to the maximum change in 'AGE'?
    * hint : maximum change of 'AGE' = max('AGE') - min('AGE')
4. **What's your take on treating age simply as a numeric variable?**
    * Do you have any other ideas on how age could be preprocessed before using it as an input in the model?

````python
####################
## Your code here ##
####################

# AGE에 대한 회귀 계수 및 표준편차 계산
beta_age = coefficients['age']
std_age = x_train['age'].std()
max_age_change = x_train['age'].max() - x_train['age'].min()

# 결과 출력
print("Y의 변화 (AGE가 1 증가 시):", beta_age)
print("AGE의 표준편차만큼 Y의 변화:", beta_age * std_age)
print("AGE의 최대 변화 시 Y의 변화:", beta_age * max_age_change)


# AGE가 1 증가할 때, 표준편차만큼 증가할 때, 최대 변화만큼 증가할 때 타겟 변수(Y)에 미치는 변화를 계산 및 해석
````
